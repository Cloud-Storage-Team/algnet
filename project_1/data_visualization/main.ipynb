# Notebook for NetworkSimulator's data visualization

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Number of packets stuck in the queue

Data produced by 30 second simulation. Servers: {10, 4, 20}

```python
iterations = []
mean_cwnd_sizes = []
queue_size = []

with open("../data/30_sec_queue_size.txt", 'r') as file:
    for line in file:
        line = line.strip()
        iteration, value, mean_cwnd_size = line.split(" ")
        iterations.append(int(iteration))
        mean_cwnd_sizes.append(int(mean_cwnd_size.strip()))  
        queue_size.append(int(value.strip()))   

print("Iteration:", iterations)
print("CWND sizes:", mean_cwnd_sizes)
print("Values:", queue_size)
```

```python
plt.title("Queue size")
plt.ylabel("Queue size, packets")
plt.xlabel("Simulation iteration")
plt.plot(iterations, queue_size, color='red')

plt.show()
```

```python
plt.title("Queue size")
plt.ylabel("Queue size, packets")
plt.xlabel("Mean CWND size, packets")
plt.plot(mean_cwnd_sizes, queue_size, color='green')

plt.show()
```
