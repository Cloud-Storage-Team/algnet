# Notebook for NetworkSimulator's data visualization

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Number of packets stuck in the queue

Data produced by 1 second simulation.

```python
iteration_index = []
queue_size = []

with open("../data/1_sec_queue_size.txt", 'r') as file:
    for line in file:
        line = line.strip()
        index, value = line.split(':')
        iteration_index.append(int(index.strip()))  
        queue_size.append(int(value.strip()))   

print("Indices:", iteration_index)
print("Values:", queue_size)
```

```python
plt.title("Queue size")
plt.ylabel("Queue size, packets")
plt.xlabel("Simulation iteration")
plt.plot(iteration_index, queue_size, color='red')

plt.show()
```
