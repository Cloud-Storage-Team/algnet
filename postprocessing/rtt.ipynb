```python
import numpy as np
import matplotlib.pyplot as plt
```

2. Run 2 simulations for incast_simulation-140 and incast_simulation-150 with output directories incast-140 and incast-150
3. Execute this notebook.

```python
min_rtt = 12450
bottleneck_rate = 50
```

```python
for N in [140,150]:
    fig, ax = plt.subplots(dpi=100)
    for n in range(N):
        ax.set_title(f'N={N}:')
        rtt = np.loadtxt(f'../build/incast-{N}/rtt/flow_conn{n+1}.txt')
        ax.plot(rtt[:,0],rtt[:,1]-min_rtt,alpha=0.5)
    bottleneck_queue = np.loadtxt(f'../build/incast-{N}/queue_size/link{2*N+1}.csv', delimiter=',', skiprows=1, usecols=(0,1))    
    ax.plot(bottleneck_queue[:,0],bottleneck_queue[:,1]*8/bottleneck_rate, c='k')
```
